```python
import pandas as pd
import numpy as np
from langdetect import detect
```

This is the original dataset that we will use for this project

```python
df = pd.read_csv("/content/drive/MyDrive/Food_ingredient_dataset/IndianFoodDatasetCSV.csv")
```

```python
df.head()
```

Below, we will remove the rows with nan in them

```python
df=df.dropna() #removing NA's
df.shape
```

```python
df = df.reset_index()
df = df.drop(['index'], axis=1)
```

```python
def detect_lang(text):   #detecting hindi language
    return detect(text)
```

Now, we will also remove those rows where the TranslatedIngredients column has hindi letters

```python
df_filtered = pd.DataFrame(columns = df.columns)
for i in range(len(df)):
  lang = detect(df['TranslatedIngredients'][i])
  if lang=='en':
    df_filtered = df_filtered.append(df.iloc[i], ignore_index = True)
```

```python
len(df_filtered)
```

```python
columns = df_filtered.columns
print(columns)
```

This is the library that we will use to parse ingredients

```python
from PyIng import parse_ingredients
```

```python
for i in range(len(df_filtered)):
  temp_ingreds = df_filtered['TranslatedIngredients'][i].split(',')
  parsed_ingredients = parse_ingredients(temp_ingreds)
  l = []
  if type(parsed_ingredients) == dict:
    l.append(parsed_ingredients['name'])
  else:
    for j in range(len(parsed_ingredients)):
      l.append(parsed_ingredients[j]['name'])
  df_filtered['TranslatedIngredients'][i] = l
```

Below, we will show how does parsed ingredient looks like
